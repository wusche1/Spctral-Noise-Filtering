```python
import numpy as np
import matplotlib.pyplot as plt

#List ofimport numpy as np
import matplotlib.pyplot as plt
#List of Gaussian random numbers
x = np.random.normal(0, 1, 1000000)
n_bins = 1000
#Create a 2x2 subplot grid
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
#Set text size for all text in the plot
text_size = 15
plt.rcParams.update({'font.size': text_size})
#Plot the normal trajectory
axs[0, 0].plot(x, '.', markersize=1)
axs[0, 0].axhline(y=0, color='black', linestyle='-', label='Mean = 0', xmin=0, xmax=1)
axs[0, 0].set_title('Trajectory')
axs[0, 0].set_xlabel('Time')
axs[0, 0].set_ylabel('Value')
#set lims
axs[0, 0].set_xlim([0, 1000000])
axs[0, 0].set_ylim([-5,5])
axs[0, 0].legend(loc='upper right')
#Plot the histogram of the normal trajectory
axs[0, 1].hist(x, bins=n_bins, color='c', density=True)
v = np.linspace(-4, 4, n_bins)
axs[0, 1].plot(v, 1/(np.sqrt(2*np.pi))*np.exp(-(v**2)/2), linewidth=2, color='k', linestyle='--', alpha=0.6, label=r'$\frac{1}{\sqrt{2\pi}}e^{-\frac{x^2}{2}}$')
axs[0, 1].axvline(x=0, color='black', linestyle='-', label='Mean=0')
axs[0, 1].set_title('Position Histogram')
axs[0, 1].set_xlabel('Value')
axs[0, 1].set_ylabel('Frequency')
axs[0, 1].legend(loc='upper right')
#set lims
axs[0, 1].set_xlim([-5, 5])
#Calculate the power spectrum
PSD = np.abs(np.fft.fft(x))**2/len(x)
freq = np.fft.fftfreq(len(x))
#Plot the PSD trajectory
axs[1, 0].plot(freq[:len(freq)//2], PSD[:len(PSD)//2], '.', markersize=1)
axs[1, 0].axhline(y=1, color='black', linestyle='-', label='Mean = 1')
axs[1, 0].set_title('Power Spectral Density')
axs[1, 0].set_xlabel('Frequency')
axs[1, 0].set_ylabel('Power')
axs[1, 0].legend(loc='upper right')
#set lims
axs[1, 0].set_xlim([0, 0.5])
axs[1, 0].set_ylim([0, 10])
#Plot the histogram of the PSD
axs[1, 1].hist(PSD, bins=n_bins, color='c', density=True)
v = np.linspace(0, 10, n_bins)
axs[1, 1].plot(v, np.exp(-v), linewidth=2, color='k', linestyle='--', alpha=0.6, label=r'$e^{-x}$')
axs[1, 1].axvline(x=1, color='black', linestyle='-', label='Mean=1')
axs[1, 1].set_title('Power Spectral Density Histogram')
axs[1, 1].set_xlabel('Power')
axs[1, 1].set_ylabel('Frequency')
axs[1, 1].legend(loc='upper right')
#set lims
axs[1, 1].set_xlim([0, 10])



#Adjust the spacing between subplots
plt.tight_layout()

#Save the plot as an SVG file
#plt.savefig('gaussian_random_numbers_analysis.png', format='png')
plt.show()
```
