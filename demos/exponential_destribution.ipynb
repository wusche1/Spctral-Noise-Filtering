```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#list of gaussian random numbers
x = np.random.normal(0,1,100000)
#plot values as histogram
plt.hist(x, bins=100, color='c',density=True)
#plot analytic pdf
v = np.linspace(-4,4,100)
plt.plot(v, 1/(np.sqrt(2*np.pi))*np.exp(-(v**2)/2), linewidth=4, color='k',linestyle='--', alpha=0.6)

plt.title('Gaussian Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
#now calculate the power spectrum
PSD = np.abs(np.fft.fft(x))**2/len(x)
#plot values as histogram
plt.hist(PSD, bins=100, color='c',density=True)
#plot analytic eponential pdf
v = np.linspace(0,10,100)
plt.plot(v, np.exp(-v), linewidth=4, color='k',linestyle='--', alpha=0.6)

plt.title('PSD of Gaussian Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python

#get the complex phase
phases = np.angle(np.fft.fft(x))
#now plot a histogram of the phase
plt.hist(phases, bins=100, color='c',density=True)
plt.title('Phase Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
def response_function(omega):
    return 1/(1* (1j*omega)**0.1 + 1 * (1j*omega)**0.75)
```

```python
frequencies = np.linspace(-10,10,1000)
plt.plot(frequencies, np.abs(response_function(frequencies)))
plt.title('Response Function')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.show()

#now plot the phase of the response function
plt.plot(frequencies, np.angle(response_function(frequencies)))
plt.title('Response Function')
plt.xlabel('Frequency')
plt.ylabel('Phase')
plt.show()
```
